# 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. 
In case if any product has less than 3 pages in search
results then scrape all the products available under that product name. Details to be scraped are: "Brand
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
urls = []
for j in range(3):
    titles = browser.find_elements_by_xpath("//a[@class='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal']")
    for i in titles:
        urls.append(i.get_attribute("href"))
    browser.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(5)
    next_button = browser.find_element_by_xpath("//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
    next_button.click()
```

```python
brand_name = []
product_name = []
price_product = []
exchange_return = []
delivery_expected = []
availability_product = []
```

```python
for i in urls:
    browser.get(i)
    try:
        brand = browser.find_element_by_xpath("//div[@class='a-expander-content a-expander-section-content a-section-expander-inner']")
        brand = brand.text.split('Brand')
        brand = brand[1].split('\n')[0]
        brand_name.append(brand[1:])
        #print(brand_name)
    except:
        try:
            brand = browser.find_element_by_xpath("//div[@id='productOverview_feature_div']")
            brand = brand.text.split('Brand')
            brand = brand[1].split('\n')[0]
            brand_name.append(brand[1:])
            #print(brand_name)
        except:
            brand_name.append('-')
            #print(brand_name)

    try:
        product = browser.find_element_by_xpath("//div[@id='titleSection']")
        product_name.append(product.text)
        #print(product_name)
    except:
        product_name.append('-')
        #print(product_name)

    try:
        price = browser.find_element_by_xpath("//div[@id='apex_desktop']//span[@class='a-price aok-align-center priceToPay']")
        price_product.append(price.text.split('\n')[0])
        #print(price_product)
    except:
        try:
            price = browser.find_element_by_xpath("//div[@id='apex_desktop']//span[@class='a-price a-text-price a-size-medium apexPriceToPay']")
            price_product.append(price.text.split('\n')[0])
            #print(price_product)
        except:
            price_product.append('-')
            #print(price_product)

    try:
        exchange = browser.find_element_by_xpath("//div[@id='maxBuyBackDiscountSection']")
        exchange_return.append(exchange.text)
        #print(exchange_return)
    except:
        try:
            exchange = browser.find_element_by_xpath("//div[@id='RETURNS_POLICY']")
            exchange_return.append(exchange.text)
            #print(exchange_return)
        except:
            exchange_return.append('-')
            #print(exchange_return)

    try:
        delivery = browser.find_element_by_xpath("//div[@id='ddmDeliveryMessage']")
        delivery = delivery.text
        delivery_expected.append(delivery.split('Details')[0])
        #print(delivery_expected)
    except:
        delivery_expected.append('-')
        #print(delivery_expected)

    try:
        availability = browser.find_element_by_xpath("//div[@id='availabilityInsideBuyBox_feature_div']")
        availability_product.append(availability.text)
        #print(availability_product)
    except:
        availability_product.append('-')
        #print(availability_product)
```

```python
amazon = pd.DataFrame()
amazon["Brand Name"] = brand_name
amazon["Name of the Product"] = product_name
amazon["Price"] = price_product
amazon["Return/Exchange"] = exchange_return
amazon["Expected Delivery"] = delivery_expected
amazon["Availability"] = availability_product
amazon["Product URL"] = urls
amazon
```

# 3.Write a python program to access the search bar and search button on images.google.com
and scrape 10 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.

```python
driver = webdriver.Chrome('chromedriver.exe')
list = ['fruits', 'cars','Machine Learning', 'Guitar', 'Cakes']
img_url=[]
keyword=[] 
for i in list:
    driver.get('https://images.google.com')
    search = driver.find_element_by_xpath("//input[@class='gLFyf gsfi']")
    search.send_keys(i)
    search.send_keys(u'\ue007')
    image = driver.find_elements_by_xpath("//div[@class='bRMDJf islir']/img")
    counter=0
    for j in image:
        if counter==10:
            break
        img_url.append(j.get_attribute('src'))
        keyword.append(i)
        counter+=1
print(len(keyword),len(img_url))
image_google = pd.DataFrame()
image_google['Keyword'] = keyword
image_google['image url'] = img_url
image_google
```

```python
image_google.to_csv('google_scrape.csv',sep='\t')
driver.close()
```

```python
search_field = browser.find_element_by_xpath("//button[@class='icon--hamburger']")
search_field.click()
billionaires = browser.find_element_by_xpath("//li[@class='header__channel header__color--centennial-silver header__hoverable']")
from selenium.webdriver.common.action_chains import ActionChains
a = ActionChains(browser)
a.move_to_element(billionaires).perform()
billionaire = browser.find_element_by_xpath("/html/body/div[1]/header/nav/div[3]/ul/li[1]/div[2]/ul/li[2]/a")
billionaire.click()
```

```python
for j in range(13):
    time.sleep(2)
    ranks=browser.find_elements_by_xpath("//div[@class='rank']")
    for i in ranks:
        rank.append(i.text)
    names=browser.find_elements_by_xpath("//div[@class='personName']")
    for i in names:
        name.append(i.text)
    worth=browser.find_elements_by_xpath("//div[@class='netWorth']")
    for i in worth:
        net_worth.append(i.text)
    ages=browser.find_elements_by_xpath("//div[@class='age']")
    for i in ages:
        age.append(i.text)
    country=browser.find_elements_by_xpath("//div[@class='countryOfCitizenship']")
    for i in country:
        citizenship.append(i.text)
    sources=browser.find_elements_by_xpath("//div[@class='source']")
    for i in sources:
        source.append(i.text)
    categories=browser.find_elements_by_xpath("//div[@class='category']")
    for i in categories:
        industry.append(i.text)
    next=browser.find_element_by_xpath("//div[@class='next-page']")
    next.click()
```

# 10. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in
“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews,
overall reviews, privates from price, dorms from price, facilities and property description.

```python
browser = webdriver.Chrome('chromedriver.exe')
browser.get('https://www.hostelworld.com/')
```

```python
from selenium.webdriver.common.action_chains import ActionChains
a = ActionChains(browser)
search = browser.find_element_by_xpath("//input[@id='location-text-input-field']")
a.move_to_element(search).perform()
search.click()
time.sleep(5)
```

```python
search_input=browser.find_element_by_xpath("//input[@id='search-input-field']")
search_input.send_keys('London')
time.sleep(5)
search_selection=browser.find_element_by_xpath("//div[@class='label']")
#search_selection=browser.find_element_by_xpath("//li[@tabindex='-1']")
search_selection.click()
```

```python
button = browser.find_element_by_xpath("/html/body/div[3]/div/div/div[1]/div[1]/div/div[2]/div[4]/div/div[2]/div/div[5]/button")
button.click()
```

```python
urls=[]
distance_city=[]
prices_col=[]
privates_from = []
dorms_from = []
total_reviews = []
overall_reviews = []
ratings = []
description=[]
review=[]
hostel_name=[]
facilities=[]
```

```python
from selenium.webdriver.common.keys import Keys
html = browser.find_element_by_tag_name('html')
for z in range(2):
    time.sleep(10)
    hostels = browser.find_elements_by_xpath("//h2[@class='title title-6']/a")
    for i in hostels:
        urls.append(i.get_attribute("href"))
    prices=browser.find_elements_by_xpath("//div[@class='price-col']")
    for i in prices:
        prices_col.append(i.text.replace('Dorms From\n',''))
    distance=browser.find_elements_by_xpath("//span[@class='description']")
    for i in distance:
        distance_city.append(i.text)
    html.send_keys(Keys.END)
    time.sleep(1)
    #next = browser.find_element_by_xpath("//div[@class='pagination-item pagination-next']")
    #next.click()
print(len(urls),len(distance_city))
```

```python
for j in range(len(distance_city)):
    distance_city[j]=distance_city[j].replace('Hostel -','')
```

```python
for i in range(len(prices_col)):
    prices_col[i]=prices_col[i].replace('Privates From\n','')
```

```python
privates_from = prices_col[0::2]
for i in range(len(privates_from)):
    if privates_from[i].count('Rs')==2:
        privates_from[i]=privates_from[i].split(' ')[1]
len(privates_from)
```

```python
dorms_from = prices_col[1::2]
for i in range(len(dorms_from)):
    if dorms_from[i].count('Rs')==2:
        dorms_from[i]=dorms_from[i].split(' ')[1]
len(dorms_from)
```

```python
for i in urls:
    browser.get(i)
    time.sleep(5)
    hostel_name.append(browser.find_element_by_xpath("//div[@class='title-2']").text)
    browser.execute_script("window.scrollTo(0, window.scrollY + 500)")
    try:
        show_more=browser.find_element_by_xpath("//a[@rel='toggle-collapse']")
        show_more.click()
    except:
        pass
    description.append(browser.find_element_by_xpath("//div[@class='description-container']").text.replace("\n",' '))
    try:
        ratings.append(browser.find_element_by_xpath("//div[@class='score orange big']").text)
    except:
        ratings.append('-')
    try:
        review=browser.find_element_by_xpath("//div[@class='summary orange']").text.split('\n')
    except:
        review=['-','-']    
    total_reviews.append(review[1])
    overall_reviews.append(review[0])
    facilities.append(browser.find_element_by_xpath("//div[@class='facilities-container']").text.replace('\n',' '))
```

```python
for i in range(len(total_reviews)):
    total_reviews[i]=total_reviews[i].replace('Total Reviews','')
```

```python
for i in range(len(description)):
    description[i]=description[i].replace('Property Description','')
```

```python
for i in range(len(facilities)):
    facilities[i]=facilities[i].replace('Facilities ','')
```

```python
hostel = pd.DataFrame()
hostel['HOSTEL NAME']=hostel_name
hostel['DISTANCE FROM CITY CENTRE']=distance_city
hostel['RATINGS']=ratings
hostel['TOTAL REVIEWS']=total_reviews
hostel['OVERALL REVIEW']=overall_reviews
hostel['PRIVATES FROM PRICE']=privates_from
hostel['DORMS FROM PRICE']=dorms_from
hostel['FACILITIES']=facilities
hostel['PROPERTY DESCRIPTION']=description

hostel
```
