# 2. Scrape the details team India’s international fixtures from bcci.tv.

```python
browser = webdriver.Chrome('chromedriver.exe')
```

```python
browser.get('https://www.bcci.tv/')
browser.maximize_window()
time.sleep(5)
```

```python
try:
    accept_button = browser.find_element_by_xpath("//button[@class='cookie__accept btn btn-primary']")
    accept_button.click()
except:
    pass
```

```python
international_button = browser.find_element_by_xpath("//a[@href='https://www.bcci.tv/international/fixtures']")
international_button.click()
```

```python
time.sleep(5)
```

```python
fixture_cards = browser.find_elements_by_xpath("//div[@class='fixture-card-top']")
fixture = []
for i in fixture_cards:
    fixture.append(i.text.split('\n'))
fixture
```

```python
series = []
date = []
matchtime = []

for i in fixture:
    series.append(i[0])
    date.append(i[1])
    matchtime.append(i[2])
```

```python
matches = browser.find_elements_by_xpath("//div[@class='fix-place ng-binding ng-scope']")
match=[]
for i in matches:
    match.append(i.text.split('\n'))
```

```python
match
```

```python
title=[]
location=[]
for i in match:
    if str(i).find('-')!=-1:
        title.append(str(i).split('-')[0])
        location.append(str(i).split('-')[1])
    else:
        title.append('-')
        location.append(i)
```

```python
for j in range(0,len(title)):
    title[j]=title[j].replace('[','')
```

```python
bcci = pd.DataFrame()
bcci['Match Title'] = title
bcci['Series'] = series
bcci['Place'] = location
bcci['Date'] = date
bcci['Time'] = matchtime
bcci
```

```python
browser.close()
```
